### Randičev indeks

```python
def randic(graf):
    '''vrne randicev indeks za nek graf'''
    vsota = 0
    povezave = graf.edges()
    stopnje = graf.degree()
    for edge in povezave:
        u, v,_ = edge
        d_u = stopnje[u]
        d_v = stopnje[v]
        vsota += 1/((d_u * d_v)**(1/2))
    return(vsota)
```
